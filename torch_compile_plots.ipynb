```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('torch_compile.csv')
df['name'] = ['Torch Compile Base', 'Torch Compile\nReduce Overhead',
              'Baseline (no Torch Compile)', 'Torch Compile\nEager Mode',]
```

```python
df['percent_speedup_tot_time'] = 100*(df.loc[(df['compiled'] == False), 'tot_time'].values - df['tot_time'].values)/df.loc[(df['compiled'] == False), 'tot_time'].values
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(df, x = 'name', y = 'tot_time', ax = ax, hue = 'compiled')
ax.set_ylabel('Total Time (s)')
ax.set_xlabel('Torch Compile Mode')
ax.set_title('Torch Compile Performance Comparison')
plt.tight_layout()
```

```python
df_subset = df.loc[(df['compiled'] == True)]
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(df_subset, x = 'name', y = 'percent_speedup_tot_time', ax = ax)
ax.set_ylabel('Percent Speedup (%)')
ax.set_xlabel('Torch Compile Mode')
ax.set_title('Torch Compile Percent Speedup Comparison')
plt.tight_layout()
```
